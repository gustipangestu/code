```python
import os

def makeDictClass(dirr):
        dictClass = {}
        i = 0
        for _i in os.listdir(dirr):
            if (os.path.isdir(os.path.join(dirr, _i)) == True):
                # print(_i)
                dictClass[i] = _i
                i = i + 1
        return dictClass

def class2id(classname, dictclass):
    id = 0
    for i in dictclass:
        if (dictclass[i] == classname):
            id = i
            break
    return id

dirr = "../../dataset/WLASL100//"
_map_class = makeDictClass(dirr=dirr)
print("lennnn", len(_map_class))
for i in _map_class:
    print(_map_class[i])
# _classId = class2id(classname=_map_class[101], dictclass=_map_class)
# print(_classId)
```
